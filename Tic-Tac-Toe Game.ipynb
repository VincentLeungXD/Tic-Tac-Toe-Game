```python
from IPython.display import clear_output
import random

#test_board=['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']

def display_board(board):
    clear_output()
    print(' '+board[7]+'  |'+board[8]+'  |'+board[9])
    print('  - +-- + -')
    print(' '+board[4]+'  |'+board[5]+'  |'+board[6])
    print('  - +-- + -')
    print(' '+board[1]+'  |'+board[2]+'  |'+board[3])

# Choosing who gets to go first
def player_order():
    if random.randint(0,1) == 0:
        return 'Player 1'
    else: 
        return 'Player 2'


def player_marker():
    choice = ' '
    while choice not in ['X','O']:
        choice = input("Player 1, please choose marker X or O: ").upper()
        if choice not in ['X','O']:
            print("Sorry, invalid choice. Please choose marker X or O.")
    if choice == 'X':
        return ('X','O')
    else:
        return ('O','X')
        
def placement(board,marker,user_placement):
    if board[user_placement]==' ':
        board[user_placement]=marker

def check_placement(board,user_placement):
    return board[user_placement] == ' '

def user_choice(board):
    user_placement=0
    while user_placement not in range(1,10) or not check_placement(board,user_placement):
        user_placement = int(input("Choose your next position. (1-9) "))
    return user_placement

def winner(board,marker):
    return((board[7] == board[8] == board[9] !=' ') or
    (board[4] == board[5] == board[6] !=' ') or
    (board[1] == board[2] == board[3] !=' ') or 
    (board[1] == board[4] == board[7] !=' ') or 
    (board[2] == board[5] == board[8] !=' ') or 
    (board[3] == board[6] == board[9] !=' ') or 
    (board[3] == board[5] == board[7] !=' ') or
    (board[1] == board[5] == board[9] !=' '))

def stalemate(board):
    for i in range(1,10):
        if check_placement(board,i):
            return False
    return True

def play_again():
    another_game=input("Do you want to play again? (Y/N) ")
    if another_game.upper() == 'Y':
        return True
        
#Piecing it all together

while True:
    theBoard = [' ']*10
    display_board(theBoard)
    player_1,player_2 = player_marker()
    turn = player_order()
    print(turn + " will go first!" )
    response = input("Are you ready to start the game? (Y/N)")
    if response.upper() == 'Y':
        game = True
    else:
        game = False
        
#Player 1's Turn
    while game:
        if turn == 'Player 1':
            display_board(theBoard)
            user_placement=user_choice(theBoard)
            placement(theBoard,player_1,user_placement)
    
            if winner(theBoard,player_1):
                display_board(theBoard)
                print('Game Over. Player 1 has won!')
                game = False
            else:
                if stalemate(theBoard):
                    display_board(theBoard)
                    print("It's a Tie!")
                    break
                else:
                    turn='Player 2'

#Player 2's Turn
        if turn == 'Player 2':
            display_board(theBoard)
            user_placement=user_choice(theBoard)
            placement(theBoard,player_2,user_placement)
    
            if winner(theBoard,player_2):
                display_board(theBoard)
                print('Game Over. Player 2 has won!')
                game = False
            else:
                if stalemate(theBoard):
                    display_board(theBoard)
                    print("It's a Tie!")
                    break
                else:
                    turn='Player 1'
       
    if not play_again():
        break
    

    
    
    
```
